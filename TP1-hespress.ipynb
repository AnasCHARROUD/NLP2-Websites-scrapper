```python
from bs4 import BeautifulSoup as bs
import requests

link = 'https://www.hespress.com'

posts_links = []
for i in range(1,20):
    request = requests.get(link+"/politique/index."+str(i)+".html")
    soup    = bs(request.text,features="lxml")
    posts   = soup.find_all('h2',attrs={"class":"section_title"})
    for post in posts:
        post_link = bs(str(post),features='lxml').find_all('a')
        posts_links.append(post_link[0]['href'])
all_comments = []
for post in posts_links:
    print(post)
    request = requests.get(link+post)
    soup = bs(request.text,features="lxml")
    comments = soup.find_all('div',attrs={'class':'comment_text'})
    for comment in comments:
        all_comments.append(comment.text)
with open('politicalComments.cvs', 'w',encoding='utf-8') as f:
    for comment in all_comments:
        f.write("%s\n" % comment)
print(len(all_comments))
```

```python
#importation des packages  
#==================================================
from nltk.corpus import stopwords 
from nltk.tokenize import WordPunctTokenizer
from nltk.stem.porter import PorterStemmer
from nltk.tokenize import RegexpTokenizer 
import csv
import re
#==================================================

#sotpwords
#==================================================
#nltk.download('stopwords')
myfile1 = open("stopwordsar.txt",encoding = "UTF-8")
stopwords_ar = myfile1.read()
#==================================================

stemmer =PorterStemmer()



#importation de notre commantaires hespress
#===============================================
myfile = open("pl.txt",encoding = "UTF-8")
text = myfile.read()
#===============================================


#netoyage des donnes.
#===============================================
text = re.sub('[^ء-ي]+',',',text)
#===============================================


#tokinze
#===============================================
tokenizer=RegexpTokenizer(r'\w+')
tokens=tokenizer.tokenize(text)
#===============================================


#elimination de stopwords
#===============================================
tokens_clean=[t for t in tokens if t not in stopwords_ar ]
#===============================================


#stemmer
#===============================================
stems=[stemmer.stem(t) for t in tokens_clean]
#===============================================









```
